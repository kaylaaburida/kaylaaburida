Basically `melt` stacks our data vertically. `seaborn` can then use `variable` column to assign groups (hues) and the `value` column to plot the values. 


````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, nrows=1, figsize=(15, 4), sharey=True, sharex=True)

with gw.open(r'\\apporto.com\dfs\GWU\Users\kaylaaburida_gwu\Downloads\assignment (1)\data\data1\rgb.tif') as src:
    df = src.gw.extract(train)
    
#print(df.head)

df_melt = {1: 'blue',
           2: 'green',
           3: 'red'}
           #, 4: 'nir'}
df.rename(columns= df_melt, inplace=True)
df = pd.melt(df, id_vars = 'name', value_vars= ['blue', 'green', 'red'])

unique_ids = df['name'].unique()
print(len(unique_ids))
axes = [ax1, ax2, ax3, ax4]

for ax, name_value in zip(axes, unique_ids):
    print(name_value)
    subset= df[df['name']== name_value]
    sns.histplot(data= subset, x= 'value', ax=ax, hue='variable', element='bars')
    ax.set_title(name_value)
    
#data.melt(value_vars=['blue','green','red'])
# YOUR CODE HERE
#raise NotImplementedError()
````

Use `geowombat` to train and predict a landcover map using [random forest classifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html) 

For this subtask:
- Set up a `pipeline` called `pipe` that creates a workflow with the following steps [docs](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html)
     - Applies sklearns `PCA` to reduce the dimensionality of the data to `2` dimensions  [docs](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html)
     - Applies the `RandomForestClassifier` classifier [docs](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html)
     - to predict `y` a landcover raster classification

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from geowombat.ml import  fit_predict

pipe = Pipeline([ ('pca', PCA(n_components=2)),
                 ('rf', RandomForestClassifier(n_estimators=100,
                                               random_state=0))])

with gw.open([r'\\apporto.com\dfs\GWU\Users\kaylaaburida_gwu\Downloads\assignment (1)\data\data1\my_rgbn_B02.tiff' ,
          r'\\apporto.com\dfs\GWU\Users\kaylaaburida_gwu\Downloads\assignment (1)\data\data1\my_rgbn_B03.tiff' ,
          r'\\apporto.com\dfs\GWU\Users\kaylaaburida_gwu\Downloads\assignment (1)\data\data1\my_rgbn_B04.tiff'],stack_dim='band',
          band_names= ['blue', 'green', 'red']) as src:
    y= fit_predict(src, pipe, train, col = 'class')

# YOUR CODE HERE
#raise NotImplementedError()
````
